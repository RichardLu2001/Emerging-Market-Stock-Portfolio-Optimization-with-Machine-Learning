```python
import pandas as pd

df = pd.read_excel("000300closeweight.xlsx",dtype={'成份券代码Constituent Code': str})

df.head()
```

```python
df['location'] = df['交易所英文名称Exchange(Eng)'].apply(lambda x: 1 if x=='Shenzhen Stock Exchange' else 0)
df.head()
```

```python
import yfinance as yf
```

```python
sample_data = df[['成份券代码Constituent Code', 'location']].copy()

data_frames = []
sample_data = sample_data.rename(columns={'成份券代码Constituent Code': 'StockCode'})

for row in sample_data.itertuples():
    stock_code = str(row.StockCode)  
    suffix = '.SZ' if row.location == 1 else '.SS' 
    ticker = f"{stock_code}{suffix}"
    
    try:
        stock_data = yf.download(ticker, start="2001-01-01", end="2024-11-12")['Adj Close']
        stock_data = stock_data.rename(ticker)
        data_frames.append(stock_data)
    except Exception as e:
        print(f"Could not download data for stock code {ticker}: {e}")
```

```python
combined_data = pd.concat(data_frames, axis=1)

# Display a sample of the data for verification
combined_data
```

```python
combined_data.to_csv('300.csv')
```
